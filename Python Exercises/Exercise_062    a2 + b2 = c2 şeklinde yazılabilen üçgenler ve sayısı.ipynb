```python
liste = []
n = int(input("Please enter a number: "))
for c in range(1, n+1) :
  for a in range(1, n+1) :
    for b in range(1, n+1) :
      if a ** 2 + b ** 2 == c ** 2 :
        print((a, b, c), end = " ")
        
        liste.append((a, b, c))
print(f"\nThere are {len(liste)} square triples that satisfy the condition for n = {n}.")
```

```python
# Alternative

n = int(input("Please enter a number: "))
for c in range(1, n+1) :
  for a in range(1, n+1) :
    for b in range(1, n+1) :
      if pow(a, 2) + pow(b, 2) == pow(c, 2) :
        print((a, b, c), end = " ")
```
