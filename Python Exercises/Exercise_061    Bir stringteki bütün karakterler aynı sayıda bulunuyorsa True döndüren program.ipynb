```python
s = "cababacbcabc"
dictionary = {} # dictionary = dict()
for i in s :
  if i not in dictionary :
    dictionary[i] = 1
  else :
    dictionary[i] += 1

print(min(dictionary.values()) == max(dictionary.values()))
```
