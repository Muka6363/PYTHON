Write a program, that takes in a positive parameter num and 
returns its multiplicative persistence, which is the number of times you must multiply the digits in num until you reach a single digit.

Pozitif bir num parametresi alan ve tek bir basamağa ulaşana kadar basamakları num cinsinden çarpmanız gereken sayı olan çarpımsal kalıcılığını döndüren bir program yazın.

image.png

```python
number = input("Please enter a number: ")
toplam = 1
count = 0

if len(number) == 1 :
  print(count)
else :
  while len(number) != 1 :
    count += 1
    for i in number :
      toplam *= int(i)
    if len(str(toplam)) == 1 :
      print(toplam)
      break
    else :
      number = str(toplam) # str(toplam) > New number  For example: 9 * 9 * 9
      toplam = 1

print(count)
```
