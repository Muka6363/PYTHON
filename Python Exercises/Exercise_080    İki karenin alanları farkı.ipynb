Imagine a circle and two squares: a smaller and a bigger one. For the smaller one, the circle is a circumcircle and for the bigger one, an incircle.
Create a program, that takes an integer (radius of the circle) and returns the difference of the areas of the two squares.

Examples

image.png

```python
n = int(input("Please enter the radius of the circle: "))
wanted = (2*n) ** 2 - (2*n / (2 ** 0.5)) ** 2
print("Difference of the areas of the two squares", round(wanted))
```
