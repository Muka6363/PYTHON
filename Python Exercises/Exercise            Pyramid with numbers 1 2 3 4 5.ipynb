Program to print half pyramid a using numbers.(User enter the numbers of rows)

Bugün farklı piramit şekilleri oluşturma ile ilgili soruları çözelim. Yarım piramit a'yı sayıları kullanarak yazdıran program.(Kullanıcı satır sayılarını girsin.)

image.png

```python
rows = int(input("Please enter the number of lines: "))

for i in range(1, rows+1) :
  for j in range(1, i+1):
    print(j, end = " ")
  print() # Her satırın sonunda alt satıra geç
```
