```python
print("---------------ATM USER INTERFACE---------------")

balance_inquiry = 1  # bakiye sorma
withdraw_money = 2   # para çekme
deposit = 3          # para yatırma
exit = 4             # işlem sonlandırma
balance = 1000
while True :
  user_choice = int(input('''Which action would you like to take:
balance_inquiry # 1
withdraw_money  # 2
deposit         # 3
exit            # 4
  '''))
  if user_choice == 1 :
    print(f"Your balance: {balance}")
    break
  elif user_choice == 2 :
    money = int(input("How much money do you want to withdraw?: "))
    balance = balance - money
    print(f"{money} ₺ was taken from your account. Your current balance: {balance} ₺")
    break
  elif user_choice == 3 :
    money = int(input("How much money do you want to deposit into your account?: "))
    balance += money
    print(f"{money} ₺ has been deposited into your account. Your current balance: {balance} ₺")
    break
  elif user_choice == 4 :
    print("We are looking forward to see you again.")
    break
```
