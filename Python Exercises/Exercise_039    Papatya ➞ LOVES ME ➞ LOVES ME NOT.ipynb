```python
n = int(input(""))
papatya = ["Loves me", "Loves me not" ]
yeni_papatya = []
if n == 1 :
  print(papatya[0].upper())
else:
  for i in range(1,n + 1) :
    if i % 2 == 1 :
      yeni_papatya += [papatya[0]]
    elif i % 2 == 0 :
      yeni_papatya += [papatya[1]]
  yeni_papatya  = yeni_papatya[:-1] + [yeni_papatya[-1].upper()]
  print(", ".join(yeni_papatya))
```
