```python
# Alternative -1

text = input("Please enter a sentence: ").split()
text_new = []
for i in text :
  if len(i) < 5 :
    text_new += [i]
  else :
    text_new += [i[::-1]]

print(" ".join(text_new))
```

```python
# Alternative - 2

def spinWords(string) :
  string = string.split()
  new_str = []
  for i in string :
    if len(i) < 5 :
      new_str += [i]
    else :
      new_str += [i[::-1]]
  return " ".join(new_str)

spinWords("Hey fellow warriors")
```
