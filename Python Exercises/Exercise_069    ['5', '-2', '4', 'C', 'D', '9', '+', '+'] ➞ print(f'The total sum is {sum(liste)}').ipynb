```python
ops = ["5", "-2", "4", "C", "D", "9", "+", "+"]
liste = []
for i in ops :
  if i == "+" :
    liste.append(liste[-1] + liste[-2])
  elif i == 'C' :
    liste.pop()
  elif i == 'D' :
    liste.append(2 * liste[-1])
  else :
    liste.append(int(i))

print(f"The total sum is: {sum(liste)}")
```
