```python
max_len = 0
max_sub = ""
s = input("")
for i in range(len(s)) :
  for j in range(1, len(s)+1):
    substring = s[i:j]
    if (substring == substring[::-1]) and (len(substring)> max_len) :
      max_sub = substring
      max_len = len(substring)

print("The longest palindromic substring: ", max_sub)
```
