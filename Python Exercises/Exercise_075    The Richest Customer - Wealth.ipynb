```python
 accounts = [[1, 2 ,3], [3, 2, 1]]
 combinedValue =[]
 count = 0
 for num in accounts :
   count = sum(num)
   combinedValue.append(count)
print(max(combinedValue))
```

```python
# Alternative

accounts = [[1 ,5], [7, 3], [3, 5]]
richest = []
order_dict = {1 : '1st', 2 : '2nd', 3 : '3rd'}
for i in accounts :
  richest.append(sum(i))
if len(set(richest)) == 1 :
  print("Both customers are richest with a wealth of {} each".format(richest[0]))
else :
  print(f"The {order_dict[int(richest.index(max(richest)) +1 )]} customer is the reachest with a wealth of {max(richest)}.")
```
