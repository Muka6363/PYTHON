A retailer is having a store-wide "buy 2, get 1 free" sale. For legal reasons, they can't charge their customers $0 for an article so a discount is applied to all products instead. For example, if a customer gets three products a, b and c:

Product A  $15.99

Product B $23.50

Product C $10.75

She gets the cheapest one for free, so she ends up paying $15.99 + $23.50 = $39.49, but what her receipt says is:

- Product A: $15.99 − Special Discount = $12.57

- Product B: $23.50 − Special Discount = $18.47

- Product C: $10.75 − Special Discount = $8.45

- Total: $39.49

Create a function that takes in a list of prices for a customer's shopping cart and returns the prices with the discount applied. Round all prices to the cent.

```python
price_list = [2.99, 5.75, 3.35, 4.99]
discount_rate = 0.0

if len(price_list) < 3 : 
  print(price_list)
else :
  discount_rate = min(price_list) / sum(price_list)

def discount(x) :
  return x - x * discount_rate
price_list = list(map(discount, price_list))
price_list = [round(num, 2) for num in price_list]
print(price_list)
```

```python
price_list, sorted_list, liste = [5.75, 14.99, 36.83, 12.15, 25.30, 5.75, 5.75, 5.75], [], []
discount_rate = 0.0

if len(price_list) < 3 : 
  print(price_list)
else :
  sorted_list = sorted(price_list)
  liste = sorted_list[::-1]
  for i in range(int(len(price_list) / 3)) :    
    liste.remove(liste[-1])

discount_rate = sum(sorted_list[:int(len(price_list) / 3)]) / sum(price_list)

def discount(x) :
  return x - x * discount_rate
price_list = list(map(discount, price_list))
price_list = [round(num, 2) for num in price_list]
print(price_list)
```
