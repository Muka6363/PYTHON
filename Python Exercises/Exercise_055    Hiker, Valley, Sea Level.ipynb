```python
steps = "DDUUUUDD"
sea_level = valley = 0
for step in steps :
  if step == "U" :
    sea_level += 1
  else :
    sea_level -= 1
  
  if step == "U" and sea_level == 0 :
    valley += 1

print(valley)
```
