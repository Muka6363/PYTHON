**Exercise - 072 | ?input() | ?for loop | ?isupper()**

Complete the solution so that the function will break up camel casing, using a space between words.

Kelimeler arasında bir boşluk kullanarak fonksiyonun deve gövdesini kırması için çözümü tamamlayın.

Example

image.png

```python
data = input("Please enter words: ")
new_data = ""
for i in data :
  if i.isupper() :
    new_data += " " + i
  else :
    new_data += i
print(new_data)
```
