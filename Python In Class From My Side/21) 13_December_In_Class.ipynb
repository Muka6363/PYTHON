```python
fruits = ["banana", "mango", "pear", "apple", "kiwi", "grape"]

count = 3

while count > 0 :
  try :
    indeks = int(input("Please enter an index number: "))
    print("Your favorite fruit is: ", fruits[indeks])
    break
  except ValueError :
    count -= 1
    print("Your input is not an integer. {} prediction(s) remaining".format(count))
  except IndexError :
    count -= 1
    print("Wrong Index Number. Try again! {} prediction(s) remaining".format(count))
  else :
    print("Valid input. There is no mistake. Congratulations!")
  finally :
    print("Our fruits are always fresh")
```

```python
def my_min(*nums):
    result = nums[0]
    for num in nums:
        if num < result:
            result = num
    return result
print(my_min(3,8,-9,0,12,1.2))
```
